## Modelization 

It seems reasonnable to fit an exponentiel curve. 

```python
def fit_expo(country='France'):
    '''Fit an exponentiel curve to the data
    country : country between quote '''
    n = sum(np.isnan(covid19[country])==False)
    x_data = range(1,n)
    y_data = covid19[country][1:n]
    log_y_data = np.log(covid19[country][1:n])
    a,b = np.polyfit(x_data, log_y_data, 1)
    y = np.exp(b) * np.exp(a*x_data)
    plt.plot(x_data, y_data, "o", label='Data')
    plt.plot(x_data, y, label='y = {:.2f}exp({:.2f}x)'.format(np.exp(b), a));
    plt.xlabel('Number of days')
    plt.legend()
    plt.title(country)
```

```python
fit_expo()
plt.ylabel('Number of death');
```

```python
fit_expo('Italie')
plt.ylabel('Number of death');
```

## Create the dashboard

```python
def fit_expo(country='France'):
    '''Fit an exponentiel curve to the data
    country : country between quote '''
    n = sum(np.isnan(covid19[country])==False)
    x_data =  range(1,n)
    y_data = covid19[country][1:n]
    log_y_data = np.log(covid19[country][1:n])
    a,b = np.polyfit(x_data, log_y_data, 1)
    y = np.exp(b) * np.exp(a*x_data)
    return(x_data, y_data, y)
    
```

```python
# france
x_data, y_data, y = fit_expo()
x_data = list(x_data)
```

```python
# italie
x_it, y_it, yit= fit_expo('Italie')
x_it = list(x_it)
```

```python
# espagne
x_es, y_es, yes = fit_expo('Espagne')
x_es= list(x_es)
```

```python
# RU
x_ru, y_ru, yru = fit_expo('Royaume-Uni')
x_ru = list(x_ru)
```

```python
fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=x_data, y=y_data,
                    name='France - Death',line = dict(color='royalblue', width=1),
                    mode='markers'))
fig.add_trace(go.Scatter(x=x_data, y=y,
                    name='France - Prediction',line = dict(color='royalblue', width=1, dash='dash'),
                    mode='lines'))

fig.add_trace(go.Scatter(x=x_it, y=y_it, line=dict(color='firebrick', width=1,
                              ),
                    mode='markers',
                    name='Italie'))
fig.add_trace(go.Scatter(x=x_it, y=yit, line=dict(color='firebrick', width=1,
                              dash='dash'),
                    mode='lines',
                    name='Italie'))


fig.add_trace(go.Scatter(x=x_es, y=y_es,line=dict(color='rgb(0,100,80)', width=1),
                    mode='markers',
                    name='Espagne'))

fig.add_trace(go.Scatter(x=x_es, y=yes,line=dict(color='rgb(0,100,80)', width=1,dash='dash'),
                    mode='lines',
                    name='Espagne'))


fig.add_trace(go.Scatter(x=x_ru, y=y_ru,line=dict(color='rgb(231,107,243)', width=1,dash='dash'),
                    mode='markers',
                    name='Royaume Uni'))
fig.add_trace(go.Scatter(x=x_ru, y=yru,line=dict(color='rgb(231,107,243)', width=1,dash='dash'),
                    mode='lines',
                    name='Royaume Uni'))

fig.show()
```
