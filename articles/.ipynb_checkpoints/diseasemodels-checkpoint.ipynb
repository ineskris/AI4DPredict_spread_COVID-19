# Disease modelling in python

This is a Jupyter notebook of python code to explore some disease models. If you are new to Jupyter, to run the code simply click in the cell and press Shift-Enter. The first two cells below must be run first for anything else to work.

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from random import randint
```

The cell above contains some commands to make everything run properly.

# Picturing an epidemic

Firstly, let's look at a discrete model. I assume 100 people in the population and $R_0=2$. Feel free to change these to see what changes

```python
N0=100             # Population size
gamma=1           # Recovery rate
R0=2              # R0
beta1=R0*gamma/N0 # Transmission coefficient

time=15
pop=np.zeros(N0)
S=np.zeros(time)
I=np.zeros(time)

for i in range(0,N0):
    pop[i]=1
    
pop[randint(1,N0)-1]=2

for step in range(0,time):
    for j in range(0,N0):
        if pop[j]==1:
            S[step]=S[step]+1
        if pop[j]==2:
            I[step]=I[step]+1
    
    newinfections=int(I[step]*R0)
    
    for j in range(0,N0):
        if pop[j]==2:
            pop[j]=3
    
    for j in range(0,newinfections):
        target=randint(1,N0)-1
        if pop[target]==1:
            pop[target]=2    
            

plt.bar(np.arange(1,time+1),I)
plt.xticks(np.arange(1,time+1))
plt.ylim(0,25)
plt.xlabel('Weeks')
plt.ylabel('No. Infected individuals')


print('There are', +int(S[time-1]), 'Susceptible individuals left')
```

Now we will model the epidemic with a continuous model using the SIR model discussed in the document. (You need to run both the next two cells).

```python
N0=100             # Population size
gamma=1           # Recovery rate
R0=2              # R0
beta1=R0*gamma/N0 # Transmission coefficient

# Function for the ODEs
def disease(x,t,beta):
    sdot=-beta*x[0]*x[1]
    idot=beta*x[0]*x[1]-gamma*x[1]
    return sdot, idot

#Baseline time course of 365 days 
ts=np.linspace(0,365,8000)

scale=10**6 # Use to make plots easier to understand
```

```python
x0=[N0-1,1]  #Initial condition
tex=np.linspace(0,15,100) #Only need a short time span for this example

xx1=odeint(disease,x0,tex,args=(beta1,)) # Run the ODEs

#Plotting commands
fig1=plt.figure()
plt.plot(tex,(xx1[:,0]),'b',label='Susceptible')
plt.plot(tex,(xx1[:,1]),'r',label='Infected')
plt.plot(tex,(N0-xx1[:,0]-xx1[:,1]),'y',label='Recovered')
plt.xlabel('Weeks')
plt.ylabel('No. individuals')
plt.legend()

fig2=plt.figure()
plt.plot(tex,beta1*xx1[:,0]/gamma)
plt.xlabel('Weeks')
plt.ylabel('$R_t$')
print('There are', int(xx1[-1,0]), 'Susceptible individuals left')
```

# A simple model for Covid-19

```python
# Updated model to use parameters LOOSELY based on Covid-19.

N0=65000000        # Population size
gamma=1/14         # Recovery rate
R0=2.5
beta1=R0*gamma/N0  # Transmission

x0=[N0-1,1]  #Initial condition assuming population of 65 million with 1 infected

xx1=odeint(disease,x0,ts,args=(beta1,))
xx2=odeint(disease,x0,ts,args=(0.75*beta1,))

fig1=plt.figure()

plt.plot(ts,(xx1[:,1]/scale),'r',label='No control')
plt.plot(ts,(xx2[:,1]/scale),'b',label='25% reduction (constant)')
plt.xlabel('Days')
plt.ylabel('No. Infected individuals (millions)')
plt.legend();
```

In this example I have used parameters *very loosely* based on Covid-19 in the UK - a population of 65 million, an $R_0$ of 2.5 and an infectious period of 2 weeks. With no control the epidemic peaks at around 15 million concurrent infections. If transmission can be reduced by 25% (for example by social distancing) then this peak is both delayed and recuced to around 8 million cases.

# Temporary control measures

What if we only wish control measures to last for a fixed time, rather than all year? Below I have set up to have a control period of 60 days starting on day 60. Play around with these values to see what happens.

```python
# Code to have temporary control measures
control_time=60    # How long control measures last for
start_cont=60      # Day control measures start
end_cont=start_cont+control_time

ta=np.linspace(0, start_cont, start_cont*10)
tb=np.linspace(start_cont,end_cont,600)
tc=np.linspace(end_cont,365,(365-end_cont)*10)
xxa=odeint(disease,x0,ta,args=(beta1,))
xxb=odeint(disease,[xxa[-1,0],xxa[-1,1]],tb,args=(0.75*beta1,))
xxc=odeint(disease,[xxb[-1,0],xxb[-1,1]],tc,args=(beta1,))

fig1=plt.figure()
plt.plot(ts,xx1[:,1]/scale,'r',label='No control')
plt.plot(ta,(xxa[:,1]/scale),'b',label='25% reduction (temporary)')
plt.plot(tb,(xxb[:,1]/scale),'b')
plt.plot(tc,(xxc[:,1]/scale),'b')
plt.xlabel('Days')
plt.ylabel('No. Infected individuals (millions)')
plt.legend(loc=1);
```

Now assume that our control measures can only be pursued for 60 days. If we launch this too early then there is almost no change to the peak of the epidemic, just a small delay. The optimal starting time seems to be between around days 120-140, just as the epdiemic is really starting to take off.

```python
control_time=60    # How long control measures last for
start_cont=120     # Day control measures start
end_cont=start_cont+control_time

ta=np.linspace(0, start_cont, start_cont*10)
tb=np.linspace(start_cont,end_cont,600)
tc=np.linspace(end_cont,365,(365-end_cont)*10)
xxa=odeint(disease,x0,ta,args=(beta1,))
xxb=odeint(disease,[xxa[-1,0],xxa[-1,1]],tb,args=(0.25*beta1,))
xxc=odeint(disease,[xxb[-1,0],xxb[-1,1]],tc,args=(beta1,))

# Second figure shows the phase plane including the nullcines
fig1=plt.figure()
#plt.plot(xx[:,0],xx[:,1],'k')
plt.plot(ts,xx1[:,1]/scale,'b',label='No control')
plt.plot(ta,(xxa[:,1]/scale),'g',label='75% reduction (60 days)')
plt.plot(tb,(xxb[:,1]/scale),'g')
plt.plot(tc,(xxc[:,1]/scale),'g')
plt.xlabel('Days')
plt.ylabel('No. Infected individuals (millions)')
plt.legend(loc=1);
```

If we increase the efficacy of our controls, we see a greater impact on the initial epidemic peak. However, because now very few individuals have been infected, once the controls are lifted a second peak occurs later in the year. That said, with good timing we can still limit both peaks to rather lower levels than previously.

# A model with low- and high-risk groups

```python
# A low-risk/high-risk group model

def disease_risk(x,t,betall,betahh,betahl):
    sldot=-betall*x[0]*x[1]-betahl*x[0]*x[3]
    ildot=betall*x[0]*x[1]+betahl*x[0]*x[3]-gamma*x[1]
    shdot=-betahh*x[2]*x[3]-betahl*x[2]*x[1]
    ihdot=betahh*x[2]*x[3]+betahl*x[2]*x[1]-gamma*x[3]
    return sldot, ildot, shdot, ihdot

ts=np.linspace(0,365,8000)
x0=[N0*0.8-1,1,N0*0.2,0]  #Initial condition

xx1=odeint(disease_risk,x0,ts,args=(beta1,beta1,beta1))
xx2=odeint(disease_risk,x0,ts,args=(0.75*beta1,beta1,0.75*beta1))
xx3=odeint(disease_risk,x0,ts,args=(beta1,0.75*beta1,0.75*beta1))
xx4=odeint(disease_risk,x0,ts,args=(0.75*beta1,0.75*beta1,0.75*beta1))

fig1=plt.figure()
plt.plot(ts,(xx1[:,1]+xx1[:,3])/scale,'r',label='No control')
plt.plot(ts,(xx3[:,1]+xx3[:,3])/scale,'g',label='Control in high-risk')
plt.plot(ts,(xx2[:,1]+xx2[:,3])/scale,'b',label='Control in low-risk')
plt.plot(ts,(xx4[:,1]+xx4[:,3])/scale,'k',label='Control in all')

plt.xlabel('Days')
plt.ylabel('No. Infected individuals (millions)')
plt.legend();

fig2=plt.figure()
plt.plot(ts,(xx1[:,3]/scale),'r',label='No control')
plt.plot(ts,(xx3[:,3]/scale),'g',label='Control in high-risk')
plt.plot(ts,(xx2[:,3]/scale),'b',label='Control in low-risk')
plt.plot(ts,(xx4[:,3]/scale),'k',label='Control in all')

plt.xlabel('Days')
plt.ylabel('No. high-risk Infected individuals (millions)')
plt.legend();
```

Here I assume that 20% of the population are high-risk individuals. How best can we proetct them? The plots show 4 cases: no control measures at all, control measures only in low-risk individuals (blue), control measure only in high-risk individuals (green) and control measures for all (black). It is clear that controlling interactions of low-risk individuals has the biggest impact. In some sense this is intuitive since this is the larger population. However, note that in the lower plot, even when we just focus on infections in high-risk individuals, there is far greater impact of putting control measures int he low-risk group.
